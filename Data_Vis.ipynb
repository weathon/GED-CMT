```python
import pylab
import numpy as np
index = 0
def get_marker():
  global index
  index+=1
  return ".,ov^<>123"[index]
def plot_box(x, y, color=None, label=None):
  pylab.yscale("log")
  ans = dict.fromkeys(set(x))
  for i in ans.keys():
    ans[i]=[]
  for i,j in zip(x,y):
    ans[i].append(j)
  mean = [np.mean(ans[i]) for i in ans.keys()]
  std = [2.8*np.std(ans[i]) for i in ans.keys()]
  pylab.errorbar(ans.keys(), mean, yerr=std, label=label, alpha=0.5, marker=get_marker(), color=color)
  pylab.ylim(0)
```
